Ques3.Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

    -Ans3:All these functions help clean messy text data by removing spaces. extracting relevant parts, formatting names, etc.
  
    a)Trim
    Its purpose is to remove extra spaces from the selected text.
    Eg: =Trim("  Deepti   Pawar    ")
    Result: "Deepti Pawar"

    b)Left
    It extracts characters from the left side.
    Eg: Left("Excel2025",5)
    -> "Excel"

    c)Right
    It extracts characters from the right side.
    Eg: Right("Excel2025",4)
    -> "2025"

    d)Mid
    It Extracts text from the middle.
    =Mid("Data2025" ,  4,4)
    ->"2025"

    e)Concat
    It combines multiple texts without space in between.
    =Concat("Deepti", " ", "Pawar")
    ->"DeeptiPawar"
